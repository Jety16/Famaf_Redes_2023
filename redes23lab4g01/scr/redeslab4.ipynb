```python
delay_simtime = data['vectime'].loc[[21]].tolist()
delay = data['vecvalue'].loc[[21]].tolist()

hops_time = data['vectime'].loc[[22]].tolist()
hops = data['vecvalue'].loc[[22]].tolist()

hops_time = list(map(float,hops_time[0].split()))
hops = list(map(float,hops[0].split()))

delay_simtime = list(map(float,delay_simtime[0].split()))
delay = list(map(float,delay[0].split()))

print(len(hops_time))
print(len(hops))

print(len(delay_simtime))
print(len(delay))

average = sum(hops) / len(hops)
print("Average Hops:", average)

average = sum(delay) / len(delay)
print("Average Delay:", average)
```

```python
# Data for the table
data = {
    "Caso 1": {
        "average hops": 3.92,
        "average delay": 51.16
    },
    "Caso 2": {
        "average hops": 4.05,
        "average delay": 0.004
    }
}
# Colors for the table rows
colors = [
    (0.537, 0.769, 0.933),  # Light Blue
    (0.247, 0.443, 0.698),  # Blue
]

# Creating the table
fig, ax = plt.subplots()

# Creating the table cells
table_data = []
for i, (row_label, row_data) in enumerate(data.items()):
    row = [row_label]
    for col_label, cell_data in row_data.items():
        row.append(cell_data)
    table_data.append(row)

# Creating the table structure
table = ax.table(cellText=table_data,
                 colLabels=["", "average hops", "average delay"],
                 cellLoc="center",
                 loc="center")

# Formatting the table
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1, 1.5)

# Hiding the table axes
ax.axis("off")

# Displaying the table
plt.show()
```

```python
file_name = 'results/exported_stads/caso2_lnkSelect.csv'
data = pd.read_csv(file_name)
data
```

```python
#Omnet guarda los valores de nuestro "vector" (lista de valores) en una celda, por lo que tengo que extraer estos valores
node_0_lnkSelected = data['vecvalue'].loc[[23]].tolist()
node_1_lnkSelected = data['vecvalue'].loc[[27]].tolist()
node_2_lnkSelected = data['vecvalue'].loc[[26]].tolist()
node_3_lnkSelected = data['vecvalue'].loc[[25]].tolist()
node_4_lnkSelected = data['vecvalue'].loc[[24]].tolist()
#node_5_lnkSelected = data['vecvalue'].loc[[30]].tolist()
node_6_lnkSelected = data['vecvalue'].loc[[22]].tolist()
node_7_lnkSelected = data['vecvalue'].loc[[21]].tolist()


# Los valores que sacamos estan como un solostring separados por comas por lo que tengo que convertirlos a listas para poder graficarlos
node_0_lnkSelected = list(map(float,node_0_lnkSelected[0].split()))
node_1_lnkSelected = list(map(float,node_1_lnkSelected[0].split()))
node_2_lnkSelected = list(map(float,node_2_lnkSelected[0].split()))
node_3_lnkSelected = list(map(float,node_3_lnkSelected[0].split()))
#node_3_lnkSelected = []
node_4_lnkSelected = list(map(float,node_4_lnkSelected[0].split()))
#node_4_lnkSelected = []
#node_5_lnkSelected = list(map(float,node_5_lnkSelected[0].split()))
node_5_lnkSelected = []
node_6_lnkSelected = list(map(float,node_6_lnkSelected[0].split()))
node_7_lnkSelected = list(map(float,node_7_lnkSelected[0].split()))

print(len(node_0_lnkSelected))
print(len(node_1_lnkSelected))
print(len(node_2_lnkSelected))
print(len(node_3_lnkSelected))
print(len(node_4_lnkSelected))
#print(len(node_5_lnkSelected))
print(len(node_6_lnkSelected))
print(len(node_7_lnkSelected))

Nodes = [node_0_lnkSelected,node_1_lnkSelected,node_2_lnkSelected,node_3_lnkSelected,node_4_lnkSelected,node_5_lnkSelected,node_6_lnkSelected,node_7_lnkSelected]

Data ={}

for i in range(8):
    lista = Nodes[i]
    cant_de_paquetes = len(lista)
    cant_de_1 = sum(lista)
    cant_de_0 = cant_de_paquetes - cant_de_1
    if(cant_de_paquetes!=0):
        pcg_0= (cant_de_0)/cant_de_paquetes
        pcg_1=(cant_de_1)/ cant_de_paquetes
    else:
        pcg_0=0
        pcg_1=0
    Data[i] = (cant_de_paquetes, cant_de_1, cant_de_0, pcg_0 *100 , pcg_1* 100)

print(Data)
```

```python
node_labels = [f"Node {i}" for i in Data.keys()]
gate_0_percentage = [x[3] for x in Data.values()]
gate_1_percentage = [x[4] for x in Data.values()]


fig, ax = plt.subplots()

index = range(len(Data))

# Plotting the horizontal bars
ax.barh(index, gate_1_percentage,  left=gate_0_percentage, color=(0.694, 0.298, 0.694), label='Gate 1', edgecolor='black', linewidth=1)
ax.barh(index, gate_0_percentage, color=(0.933, 0.537, 0.933),   label='Gate 0', edgecolor='black', linewidth=1)

# Adding labels and title
ax.set_xlabel('Percentage of Times Used')
ax.set_ylabel('Node')
ax.set_yticks(index)
ax.set_yticklabels(node_labels, weight='bold')
ax.set_title('Gate Usage by Node')
# Adjusting the x-axis limits
ax.set_xlim([0, 120])

# Adding a legend
ax.legend()

plt.tight_layout()
plt.show()
```
